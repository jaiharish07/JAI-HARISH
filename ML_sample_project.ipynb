```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
data=pd.DataFrame({"Age":[5,10,20,30,50,13,60],"salary":[0,5000,21000,13000,40000,5500,42000],"insurance":[0,0,1,0,1,0,1]}) # input datas
print(data)
x=data.drop("Age",axis=1)
y=data["Age"]
kn=KNeighborsClassifier(n_neighbors=2)
kn.fit(x,y) #fiting datas
New=pd.DataFrame({"salary":[0,15000,27000,23000,49000],"insurance":[0,0,1,1,1]}) #inserting new data
New["prediction"]=kn.predict(New) #prediction
print(New)
plt.scatter(
    data["Age"], #x-axis of the plot
    data["salary"],  #y-axis of the plot
    c=data["insurance"]   #point on plot
)
plt.show()
plt.scatter(
    New["salary"], #x-axis of the plot
    New["prediction"],  #y-axis of the plot
   #point on plot
)
plt.show()

```
